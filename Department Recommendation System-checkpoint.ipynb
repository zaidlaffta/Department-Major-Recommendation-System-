```python
#import pandas library needed
import pandas as pd
import numpy as np
from scipy.spatial.distance import euclidean, pdist, squareform
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
#read data from file
df = pd.read_csv('CleanData.csv')
#check data by printing the head just to make sure 
#We have data
print (df.head()) 
# make sure that program ID is correct in a file.
program_name = pd.read_csv('ProgramTitles.csv')
#printing section
print ("PROGRAM ID iS; ")
print (program_name) 
print ("-------------------------------")
```

```python
# merge the program title with the program id
data = pd.merge(df, program_name, on='item_id')
print ("Data after Merge is: ")
print (data.head())
print ("-------------------------------")
```

```python
#Calculate count rating of all Programs/Department
#and disply those rating ascendingly
ratingCount = data.groupby('title')['rating'].count().sort_values(ascending=False) 
print ("This is the rating count of all some of the programs:   ")
print (ratingCount)
print ("-------------------------------")
```

```python
# Calculate mean rating of all Programs/Department 
meanRating = data.groupby('title')['rating'].mean().sort_values(ascending=False)
print ("Mean rating for the Programs: ")
#display the departments ascendingly according to the mean
print (meanRating)
print ("-------------------------------")
```

```python
#Creating dataframe with 'rating' count values 
ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
ratings['num of ratings'] = pd.DataFrame(data.groupby('title')['rating'].count()) 
print ("rating for all Programs")
print (ratings['num of ratings'])
print ("-------------------------------")
```

## This is a simply way to let the user to rate this Recommendation system:
This simple progrem will take a user's rating and put it in a CSV file then compile the new ratings with all other old ratings and show you what is the total rating statistics

```python
import statistics
rating  = int (input("enter input: "))
#a loop to make sure we have the right rating entered
while rating < 0 or rating > 10:
    print ("please Enter a valid rating, it should be between 0 and 10: ")
    rating  = int (input("enter input: "))
print ("Your rating is: ", rating)

#Function to add the rating value to CSV file called Ratings
def add_rating(value):
    with open('P_ratings.csv','a') as citylist:
        citylist.write('%s\n' % value)
        citylist.close()
add_rating(rating)
df_ratings = pd.read_csv("P_ratings.csv")
type(df_ratings.index)
print ("Mean of all rating for all times is: ", df_ratings.mean())
print ("The lowest rating for all times is: ", df_ratings.min())
print ("The highest rating of all rating for all times is: ", df_ratings.max())
# show rating of this recommendatoin system using a graph
plt.title("Rating of this recommendation system")
#show as a line graph
plt.plot(df_ratings)
#show as a bar graph
df_ratings.plot(kind = 'bar')
```
